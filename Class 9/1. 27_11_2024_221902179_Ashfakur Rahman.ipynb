```python
# Q2: Write a program that takes two numbers as input and calculates the sum, difference,product, and quotient.

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

print("Sum: ", x + y)
print("Difference: ", x - y)
print("Product: ", x * y)
print("Quotient: ", x / y)

```

```python
# Q3: Create a program that takes the length and width of a rectangle as input from the userand calculates the area.

length = int(input("Enter length: "))
width = int(input("Enter width: "))
print("Area: ", length*width)
```

```python
# Q8 : Write a program that takes a list of five numbers and prints the minimum,maximum, and sum of the numbers using built-in functions.


numbers = []

for i in range(5):
    num = float(input(f"Enter number {i+1}: "))
    numbers.append(num)

min_num = min(numbers)
max_num = max(numbers)
total_sum = sum(numbers)
234
print(f"Minimum number: {min_num}")
print(f"Maximum number: {max_num}")
print(f"Sum of numbers: {total_sum}")
```
