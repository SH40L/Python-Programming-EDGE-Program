```python
# Q1: Write a program that takes an integer input from the user and checks if it is divisible by both 3 and 5. If it is, print ”Divisible by 3 and 5”; otherwise, print ”Not divisible by 3 and 5”.

num = int(input("Enter a number: "))

if(num % 3 == 0 and num % 5 == 0):
  print("Divisible by 3 and 5")
else:
  print("Not divisible by 3 and 5")
```

```python
# Q2: Create a program that takes a positive integer from the user and calculates the sum of its digits.

num = int(input("Enter a positive integer: "))

digit_sum = 0

while num > 0:
    digit_sum += num % 10
    num //= 10

# Print the sum of the digits
print(f"The sum of the digits is: {digit_sum}")
```

```python
# Q3: Develop a program that prompts the user to enter a series of numbers (the input should continue until the user enters 0). The program should then count how many of those numbers are even and how many are odd.


# Initialize counters for even and odd numbers
even_count = 0
odd_count = 0

# Loop to continuously prompt the user for numbers
while True:
    # Take an integer input from the user
    num = int(input("Enter a number (enter 0 to stop): "))

    # Check if the user entered 0 to stop the loop
    if num == 0:
        break

    # Check if the number is even or odd
    if num % 2 == 0:
        even_count += 1  # Increment the even counter
    else:
        odd_count += 1   # Increment the odd counter

# Print the results after the user enters 0
print(f"Even numbers count: {even_count}")
print(f"Odd numbers count: {odd_count}")
```
