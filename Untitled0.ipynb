```python
# Create a 1D NumPy array from a Python list [10,20,30,40]. Print the array.
import numpy as np
my_list = [10, 20, 30, 40]
my_array = np.array(my_list)
print(my_array)
```

```python
# Create a 2D NumPy array from a Python list [[1,2],[3,4],[5,6]] and print it.
my_list = [[1, 2], [3, 4], [5, 6]]
my_array = np.array(my_list)
print(my_array)
```
