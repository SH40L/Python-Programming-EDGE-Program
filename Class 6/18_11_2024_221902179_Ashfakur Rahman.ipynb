```python
def find_vowels(str_):
    vowels = "aeiouAEIOU"
    vowels_in_string = []
    for char in str_:
        if char in vowels:
            vowels_in_string.append(char)
    return vowels_in_string

str_ = "A quickly brown fox jumps over the lazy dog."
vowels = find_vowels(str_)
print("Vowels in the string:", vowels)
```

```python
def Palindrome(word):

    rw = word[::-1]

    if word == rw:
        return True
    else:
        return False

if Palindrome(12321):
    print('Palindrome')
else:
    print('Not Palindrome')
```
