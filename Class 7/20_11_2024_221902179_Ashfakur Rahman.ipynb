```python
# prompt: #Write a Python Program to get 5 Words from yhe user. Iterate that List to Display only those Word that starts with "b".

words = []
for i in range(5):
  word = input(f"Enter word {i + 1}: ")
  words.append(word)

for word in words:
  if word.startswith('b') or word.startswith('B'):
   print(word)
```

```python
# prompt: write a python program to get 5 items from user to store in a list. user should be able to remove first item of the list.

my_list = []
for i in range(5):
  item = input(f"Enter item {i + 1}: ")
  my_list.append(item)

print("Original list:", my_list)

if len(my_list) > 0:
  my_list.pop(0)
  print("List after removing the first item:", my_list)
else:
  print("List is empty, nothing to remove.")
```

```python
# prompt: write a pyrhon progam to get 5 inter numbers from user and store in a list, display that number using for loop, every number should display with its square.

numbers = []
for i in range(5):
  while True:
    try:
      number = int(input(f"Enter integer number {i + 1}: "))
      numbers.append(number)
      break
    except ValueError:
      print("Invalid input. Please enter an integer.")

for number in numbers:
  print(f"Number: {number}, Square: {number ** 2}")
```

```python
# prompt: write a python program to get 5 words from user, display that word which ends with 'n' character.

words = []
for i in range(5):
  word = input(f"Enter word {i + 1}: ")
  words.append(word)

for word in words:
  if word.endswith('n'):
    print(word)
```
