```python
class IterativeDeepening:
    def __init__(self):
        self.stack = []
        self.numberOfNodes = 0
        self.depth = 0
        self.maxDepth = 0
        self.goalFound = False

    def iterativeDeepening(self, adjacencyMatrix, destination):
        self.numberOfNodes = len(adjacencyMatrix)  # Adjust for 0-based indexing
        while not self.goalFound:
            self.depthLimitedSearch(adjacencyMatrix, 0, destination)  # Start from node 0 (0-based index)
            self.maxDepth += 1
            if self.goalFound:
                print("\nGoal Found at depth", self.maxDepth)
                return
            self.stack = []
            self.depth = 0
            self.goalFound = False  # Reset goalFound for the next iteration

    def depthLimitedSearch(self, adjacencyMatrix, source, goal):
        visited = [False] * self.numberOfNodes  # Use False for unvisited nodes
        self.stack.append(source)
        visited[source] = True
        print("\nAt Depth", self.maxDepth)

        while self.stack:
            element = self.stack[-1]
            found = False
            for destination in range(self.numberOfNodes):  # Loop through all possible nodes
                if self.depth < self.maxDepth:
                    if adjacencyMatrix[element][destination] == 1 and not visited[destination]:
                        visited[destination] = True
                        self.depth += 1
                        self.stack.append(destination)
                        print(destination, end='\t')  # Fixed the indentation here
                        if destination == goal:
                            self.goalFound = True
                            return
                        found = True
                        break

            if not found:
                self.stack.pop()
                self.depth -= 1


if __name__ == "__main__":
    try:
        print("Enter the number of nodes in the graph\n")
        number_of_nodes = int(input().strip())

        adjacency_matrix = [[0] * number_of_nodes for _ in range(number_of_nodes)]

        print("Enter the adjacency matrix\n")
        for i in range(number_of_nodes):
            row = list(map(int, input().strip().split()))
            for j in range(number_of_nodes):
                adjacency_matrix[i][j] = row[j]

        print("Enter the destination node (0-based index):\n")
        destination = int(input().strip())

        iterativeDeepening = IterativeDeepening()
        iterativeDeepening.iterativeDeepening(adjacency_matrix, destination)
    except ValueError:
        print("Wrong Input format")
```
