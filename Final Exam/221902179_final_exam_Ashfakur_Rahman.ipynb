```python
import pandas as pd
import numpy as np
```

```python
# Problem 1: Employee Salary Analysis

# Load the dataset
df = pd.read_csv("employees_large.csv")

# Compute the average salary for each department
average_salaries = df.groupby("Department")["Salary"].mean()

# Identify and print the department with the highest average salary
highest_salary_department = average_salaries.idxmax()

print("Average salaries by department:")
print(average_salaries)
print(f"Department with highest average salary: {highest_salary_department}")
```

```python
#Problem 2: Sales Analysis Using Pandas


# Load the dataset
df = pd.read_csv("sales_large.csv")

# Compute the total sales for each product in each region
total_sales = df.groupby(["Region", "Product"])["Sales"].sum()

# Identify and print the product with the highest total sales across all regions
highest_selling_product = df.groupby("Product")["Sales"].sum().idxmax()

print("Total sales for each product in each region:")
print(total_sales)
print(f"Product with highest total sales: {highest_selling_product}")
```

```python

# Define matrices A and B
A = np.array([[2, 4, 6], [1, 3, 5], [7, 9, 11]])
B = np.array([[8, 5, 2], [7, 4, 1], [6, 3, 0]])

# Perform matrix operations
sum_matrix = A + B
difference_matrix = A - B
product_matrix = A * B

print("Sum of matrices:\n", sum_matrix)
print("Difference of matrices:\n", difference_matrix)
print("Element-wise product of matrices:\n", product_matrix)
```

```python
#Problem 4: Statistical Analysis of a NumPy Array

# Generate a NumPy array of 10,000 random real-valued numbers
data = np.random.randn(10000)

# Compute mean and standard deviation
mean_value = np.mean(data)
std_dev = np.std(data)

# Normalize the data (scaling to range [0,1])
normalized_data = (data - np.min(data)) / (np.max(data) - np.min(data))

# Count how many values are greater than the mean
greater_than_mean_count = np.sum(data > mean_value)

print("Mean of the dataset:", mean_value)
print("Standard deviation of the dataset:", std_dev)
print("Number of values greater than the mean:", greater_than_mean_count)
```

```python
#Problem 5: Temperature Analysis Using List Comprehensions

# Given list of daily temperatures
temperatures = [32, 35, 38, 29, 31, 28, 27, 35, 30, 31, 29, 33, 34, 36, 37, 38, 39]

# 1. Filter temperatures above 32°C
above_32 = [temp for temp in temperatures if temp > 32]

# 2. Generate a list of tuples containing temperature and classification
temperature_classification = [
    ("Hot", temp) if temp > 35 else ("Warm", temp) if 30 <= temp <= 35 else ("Cool", temp)
    for temp in temperatures
]

# 3. Compute temperature differences from the average
average_temperature = sum(temperatures) / len(temperatures)
temperature_differences = [temp - average_temperature for temp in temperatures]

print("Temperatures above 32°C:", above_32)
print("Temperature classification:", temperature_classification)
print("Temperature differences from average:", temperature_differences)
```
