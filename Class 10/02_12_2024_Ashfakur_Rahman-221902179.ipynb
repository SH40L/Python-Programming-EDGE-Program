```python
# Write a Python Program to get 10 Numbers from user, Store in Tuple, and Display only Odd Numbers getting from the Tuple.

numbers = []

for i in range(10):
    num = int(input(f"Enter number {i+1}: "))
    numbers.append(num)

numbers_tuple = tuple(numbers)

print("Odd numbers in the tuple are:")
for num in numbers_tuple:
    if num % 2 != 0:
        print(num)
```

```python
# Write a Python Program to get 5 any Words from user, Iterate that Tuple to Display only those Word Name that ends with “n”.

words = []

for i in range(5):
    word = input(f"Enter word {i+1}: ")
    words.append(word)

words_tuple = tuple(words)

print("Words that end with 'n' are:")
for word in words_tuple:
    if word.endswith('n'):
        print(word)
```
